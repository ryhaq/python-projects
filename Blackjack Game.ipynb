This simple variation of Blackjack sees the player and AI dealer be dealt two cards each, with one of the dealer's cards being hidden. The player will choose whether to be dealt another card (hit) or to remain (stand). The dealt ace card is either 1 or 11 depending on what is most benefical to the user. The dealer will also hit or stand and whoever is closer to 21, without exceeding it, wins.

```python
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,'Queen':10, 'King':10, 'Ace':11}

class Card:
    def __init__(self,ranks,suits):
        self.suit = suits
        self.rank = ranks
        self.value = values[ranks]
        
    def __str__(self):
        return f'{self.rank} of {self.suit}'
```

```python
import random
class Deck:
    
    
    def __init__(self):
        self.all_cards = []
        for suit in suits:
            for rank in ranks:
                self.all_cards.append(Card(rank,suit))
                
    def shuffle(self):
        random.shuffle(self.all_cards)
    
    def deal(self):
        return self.all_cards.pop()
                
```

```python
class Hand:
    def __init__(self):
        self.cards = []  
        self.value = 0   
           
        
    def add_card(self,card):
        self.cards.append(card)
        self.value = self.value + card.value
        
    
    def adjust_for_ace(self,values):
        if self.value + 11 > 21:
            values['Ace'] = 1
        else:
            values['Ace'] = 11
    
        
```

```python
def hit(deck,hand):
    
    hand.adjust_for_ace(values)
    
    if hand.value < 21:
        hand.add_card(deck.deal())
    
    
```

```python
def hit_or_stand(deck,hand):
    global playing
    
    decision = ''
    while decision not in ['Hit','Stand']:
        decision = input('Hit or Stand?      ').capitalize()
    if decision == 'Hit':
        hit(deck,hand)
    else:
        playing = False
    
        
        
    
```

```python
def show_some(player,dealer):
    print("\nDealer's cards:")
    print('*hidden card*')
    for x in dealer.cards[:-1]:
        print(x)
    print("\nPlayer's cards:")
    for y in player.cards:
        print(y)
    
```

```python
def show_all(player,dealer):
    print("\nDealer's cards:")
    for x in dealer.cards:
        print(x)
    print(f'{dealer.value}')
    print("\nPlayer's cards:")
    for y in player.cards:
        print(y)
    print(f'Total: {player.value}')
    
```

```python
def player_busts(player,dealer):
    print("Player busts!")
    

def player_wins(player,dealer):
    print("Player wins!")
    

def dealer_busts(player,dealer):
    print("Dealer busts!")
    
    
def dealer_wins(player,dealer):
    print("Dealer wins!")
    
    
def push(player,dealer):
    print("Dealer and Player tie! It's a push.")
```

```python
playing = True
game_on = True
choice = ''

while game_on == True:
    
    deck = Deck()
    player = Hand()
    dealer = Hand()
    

    print('welcome to game')
    deck.shuffle()
    
    for num in range(2):
        player.add_card(deck.deal())
        dealer.add_card(deck.deal())
        
   
    
    show_some(player,dealer)
    
    while playing:
        
        hit_or_stand(deck,player)
        
        
        
        if player.value > 21:
            show_all(player,dealer)
            player_busts(player,dealer)
            break
        
        if player.value < 22:

            if dealer.value < 17:
                dealer.add_card(deck.deal())
                print('Dealer hits')
            show_all(player,dealer)



            if dealer.value > 21:
                dealer_busts(player,dealer)
                playing = False

            elif dealer.value > player.value and dealer.value < 22:
                dealer_wins(player,dealer)
                playing = False

            elif player.value > dealer.value and player.value < 22:
                player_wins(player,dealer)
                playing = False


            elif player.value == dealer.value:
                push(player,dealer)
                playing = False

                
    
    choice =''
    while choice not in ['Y','N']:
        choice = input('Play again? (Y/N)  ').capitalize()
    if choice == 'Y':
        playing = True

    else:
        game_on = False
        break
                
```
