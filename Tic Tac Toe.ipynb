```python
def x_or_o():
   
    character = ''
    is_x = False
    while character not in ['X','O']:
        character = input('Player 1, choose X or O ')
    if character == 'X':
        print('Player 1 is X\nPlayer 2 is O')
        is_x = True
        return 'X'
    else:
        print('Player 1 is O\nPlayer 2 is X')
        return 'O'
        
```

```python
def used_move(position,used):
    used.append(position)
```

```python
def user_move(used):
    
    
    is_used = True
    in_range = False
    move = ''
    
    while in_range == False or move.isdigit() == False or is_used == True:
        move = input(' enter a move 1-9 ')
        
        
       
            
        if move.isdigit() == True:
            if int(move) in range(1,10):
                    in_range = True
            if int(move) not in used:
                    is_used = False
            else:
                print('this space is already occupied')
    
    position = int(move)
    
    return position
```

```python
def change_display(character,position,row1,row2,row3):
    if position in [1,2,3]:
        row1[position-1] = character
    if position in [4,5,6]:
        row2[position-4] = character
    if position in [7,8,9]:
        row3[position-7] = character
        
    
    print(row1)
    print(row2)
    print(row3)
        
    
```
